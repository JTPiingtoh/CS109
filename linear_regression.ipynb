```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("sales_data_sample.csv")
```

```python
df.head()
```

```python
df_sales = df[["PRICEEACH", "SALES"]].sample(200)
df_sales = df_sales[df_sales.PRICEEACH < 100]
```

```python
plt.scatter(df_sales.PRICEEACH.values, df_sales.SALES.values)
```

```python
X = df_sales[["PRICEEACH"]].values
y = df_sales["SALES"].values
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(f"Y = {reg.intercept_} + {reg.coef_[0]}X")
```

```python
reg.predict(np.array([[100]]))
```
